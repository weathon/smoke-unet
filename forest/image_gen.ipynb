```python
import openai
from openai import OpenAI
from google.colab import userdata

client = OpenAI(api_key=userdata.get('key'))
import os
count = 0
for i in range(20):
  res = client.images.generate(
    model="dall-e-2",
    prompt="A panoramic view of a lush forest seen from a high vantage point, the camera is viewing the horizon in summer or fall or winter (no snow) season. Air is cler, WITH NO smoke or fog.",
    n=8,
    size="1024x1024"
  )
  for j in res.data:
    os.system('wget -O '+str(count)+'.png "'+j.url+'"')
    print(j.url)
    count+=1
```
